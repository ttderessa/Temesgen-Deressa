```python
"""
Hedonic Pricing using Object-Oriented Programming

Author: Dr.Temesgen Deressa
Date: March 26, 2026
Description:
This script demonstrates hedonic pricing by modeling a house as an object
and estimating price using both a manual function and a regression model.
"""

import numpy as np
from sklearn.linear_model import LinearRegression

# -------------------------------
# 1. Define the House class
# -------------------------------
class House:
    def __init__(self, size, rooms, air_quality, distance_to_park):
        self.size = size
        self.rooms = rooms
        self.air_quality = air_quality
        self.distance_to_park = distance_to_park

    # Manual pricing (assumed coefficients)
    def calculate_price_manual(self):
        return (
            100 * self.size +
            10000 * self.rooms +
            5000 * self.air_quality -
            2000 * self.distance_to_park
        )

    # Convert object to feature array (for ML model)
    def to_features(self):
        return [self.size, self.rooms, self.air_quality, self.distance_to_park]


# -------------------------------
# 2. Create sample data (training)
# -------------------------------
# Features: [size, rooms, air_quality, distance_to_park]
X = np.array([
    [2000, 3, 8, 0.5],
    [1800, 3, 6, 1.0],
    [2200, 4, 9, 0.2],
    [1500, 2, 5, 2.0],
    [1700, 3, 7, 1.5]
])

# Prices (simulated market prices)
y = np.array([350000, 300000, 420000, 250000, 310000])


# -------------------------------
# 3. Train hedonic pricing model
# -------------------------------
model = LinearRegression()
model.fit(X, y)

print("Learned coefficients (implicit prices):")
print(f"Size: {model.coef_[0]:.2f}")
print(f"Rooms: {model.coef_[1]:.2f}")
print(f"Air Quality: {model.coef_[2]:.2f}")
print(f"Distance to Park: {model.coef_[3]:.2f}")
print(f"Intercept: {model.intercept_:.2f}")


# -------------------------------
# 4. Create new house objects
# -------------------------------
house1 = House(2000, 3, 8, 0.5)
house2 = House(2000, 3, 5, 5)


# -------------------------------
# 5. Compare pricing methods
# -------------------------------
print("\n--- Manual Pricing ---")
print(f"House 1 Price: ${house1.calculate_price_manual():,.2f}")
print(f"House 2 Price: ${house2.calculate_price_manual():,.2f}")

print("\n--- ML Predicted Pricing (Hedonic Model) ---")
pred1 = model.predict([house1.to_features()])[0]
pred2 = model.predict([house2.to_features()])[0]

print(f"House 1 Predicted Price: ${pred1:,.2f}")
print(f"House 2 Predicted Price: ${pred2:,.2f}")
```
