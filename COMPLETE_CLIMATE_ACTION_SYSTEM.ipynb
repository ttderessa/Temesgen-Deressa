```python

# ===========================================================
# COMPLETE CLIMATE ACTION SYSTEM (SIMPLIFIED BAYESIAN VERSION)
# All Agents + Monte Carlo + Simplified Bayesian Learning
# ===========================================================
# Author: Dr. Temesgen Deressa
# Date: 2026-01-31
# Description: This system models climate dynamics, mitigation,
# adaptation, Bayesian learning, and investment strategies.
# It includes Monte Carlo simulations, scenario analysis,
# and visualization of climate risks and strategies.
# ===========================================================


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')
import json

np.random.seed(42)

# -------------------------
# 1. ORIGINAL CLIMATE AGENTS
# -------------------------

class DataCollectionAgent:
    def __init__(self, region="global"):
        self.region = region

    def run(self, state):
        print(f"📡 Collecting data for {self.region}...")

        state["raw_data"] = {
            "region": self.region,
            "timestamp": datetime.now().isoformat(),
            "ghg_emissions": {
                "CO2": 416.45,
                "CH4": 1.875,
                "N2O": 0.334
            },
            "temperature_anomaly": 1.2,
            "sea_level_rise": 3.4,
            "arctic_ice_extent": 4.21,
            "renewable_share": 29.3,
            "deforestation_rate": 10.0,
            "carbon_budget_remaining": 400
        }

        state["metadata"] = {
            "data_quality": "high",
            "collection_method": "satellite_ground_stations",
            "time_range": "2000-2024"
        }

        return state


class ClimateModelingAgent:
    def __init__(self, model_type="IPCC_AR6"):
        self.model_type = model_type

    def run(self, state):
        print("🌡️ Running climate projections...")

        raw = state["raw_data"]
        current_temp = raw["temperature_anomaly"]

        scenarios = {
            "ssp1_2.6": {"description": "Sustainable Development", "emissions_reduction": 0.7},
            "ssp2_4.5": {"description": "Middle of the Road", "emissions_reduction": 0.3},
            "ssp3_7.0": {"description": "Regional Rivalry", "emissions_reduction": -0.1},
            "ssp5_8.5": {"description": "Fossil Fuel Development", "emissions_reduction": -0.5}
        }

        projections = {}
        for scenario, params in scenarios.items():
            temp_2100 = current_temp * (1 + (0.03 * (1 - params["emissions_reduction"]) * 76))
            sea_level_2100 = raw["sea_level_rise"] * (1 + (0.02 * (1 - params["emissions_reduction"]) * 76))

            projections[scenario] = {
                "temperature_2100": round(temp_2100, 2),
                "sea_level_rise_2100": round(sea_level_2100, 1),
                "probability": max(0.1, 0.5 - abs(params["emissions_reduction"]) * 0.3)
            }

        state["projections"] = projections
        state["tipping_points"] = self._assess_tipping_points(raw)

        return state

    def _assess_tipping_points(self, data):
        thresholds = {
            "amazon_dieback": {"threshold": 3.5, "current": data["temperature_anomaly"], "risk": "medium"},
            "permafrost_melt": {"threshold": 1.5, "current": data["temperature_anomaly"], "risk": "high"},
            "ice_sheet_collapse": {"threshold": 2.0, "current": data["temperature_anomaly"], "risk": "medium"},
            "amoc_collapse": {"threshold": 4.0, "current": data["temperature_anomaly"], "risk": "low"}
        }

        for point, values in thresholds.items():
            proximity = values["current"] / values["threshold"]
            if proximity >= 0.8:
                values["risk"] = "critical"
            elif proximity >= 0.6:
                values["risk"] = "high"

        return thresholds


class ImpactAssessmentAgent:
    def run(self, state):
        print("⚡ Assessing climate impacts...")

        projections = state["projections"]["ssp2_4.5"]

        # Economic impacts
        gdp_loss_pct = min(0.3, projections["temperature_2100"] * 0.05)

        # Human impacts
        population_at_risk = {
            "heat_stress": int(3.5e9 * (projections["temperature_2100"] / 4.0)),
            "water_scarcity": int(2.0e9 * (projections["temperature_2100"] / 3.0)),
            "food_insecurity": int(800e6 * (projections["temperature_2100"] / 2.5))
        }

        # Ecological impacts
        species_loss_pct = min(0.4, projections["temperature_2100"] * 0.1)

        state["impacts"] = {
            "economic": {
                "gdp_loss_percent": round(gdp_loss_pct * 100, 1),
                "cost_of_inaction": round(20 * projections["temperature_2100"], 1),
                "climate_migration": round(140 * projections["temperature_2100"])
            },
            "human": population_at_risk,
            "ecological": {
                "species_loss_percent": round(species_loss_pct * 100, 1),
                "coral_reef_loss": round(90 * (projections["temperature_2100"] / 2.0), 1),
                "forest_loss": round(30 * (projections["temperature_2100"] / 2.5), 1)
            },
            "health": {
                "excess_deaths_per_year": int(250000 * projections["temperature_2100"]),
                "disease_spread_increase": round(15 * projections["temperature_2100"], 1)
            }
        }

        return state


class MitigationStrategyAgent:
    def __init__(self, budget_constraint=1e12):
        self.budget = budget_constraint

    def run(self, state):
        print("🌿 Developing mitigation strategies...")

        carbon_budget = state["raw_data"]["carbon_budget_remaining"]

        strategies = {
            "energy_transition": {
                "description": "Shift to renewable energy",
                "cost": 300e9,
                "emissions_reduction": 0.35,
                "timeline": "2030",
                "co_benefits": ["air_quality", "energy_security", "jobs"]
            },
            "reforestation": {
                "description": "Large-scale forest restoration",
                "cost": 50e9,
                "emissions_reduction": 0.07,
                "carbon_sequestration": 100,
                "timeline": "2050",
                "co_benefits": ["biodiversity", "water_cycle", "livelihoods"]
            },
            "carbon_capture": {
                "description": "Direct air capture technology",
                "cost": 200e9,
                "emissions_reduction": 0.15,
                "scalability": "medium",
                "timeline": "2040",
                "co_benefits": ["tech_innovation"]
            },
            "agricultural_reform": {
                "description": "Sustainable farming practices",
                "cost": 80e9,
                "emissions_reduction": 0.10,
                "timeline": "2035",
                "co_benefits": ["food_security", "soil_health"]
            }
        }

        affordable_strategies = {}
        total_cost = 0
        total_reduction = 0

        for name, strategy in strategies.items():
            if total_cost + strategy["cost"] <= self.budget:
                affordable_strategies[name] = strategy
                total_cost += strategy["cost"]
                total_reduction += strategy["emissions_reduction"]

        paris_compliance = total_reduction >= 0.45

        state["mitigation"] = {
            "strategies": affordable_strategies,
            "total_cost": total_cost,
            "total_emissions_reduction": total_reduction,
            "carbon_budget_remaining_after": carbon_budget * (1 - total_reduction),
            "paris_agreement_compatible": paris_compliance,
            "net_zero_year": 2050 if paris_compliance else 2070
        }

        return state


class AdaptationPlanningAgent:
    def run(self, state):
        print("🛡️ Planning adaptation measures...")

        impacts = state["impacts"]
        projections = state["projections"]["ssp2_4.5"]

        adaptation_measures = {
            "coastal_protection": {
                "priority": "high" if projections["sea_level_rise_2100"] > 50 else "medium",
                "cost": 150e9,
                "beneficiaries": "coastal_populations",
                "effectiveness": 0.8
            },
            "agricultural_adaptation": {
                "priority": "high",
                "cost": 60e9,
                "beneficiaries": "farmers_rural_communities",
                "effectiveness": 0.7
            },
            "water_management": {
                "priority": "medium",
                "cost": 90e9,
                "beneficiaries": "arid_region_populations",
                "effectiveness": 0.75
            }
        }

        for measure in adaptation_measures.values():
            measure["cost_benefit_ratio"] = round(
                measure["effectiveness"] / (measure["cost"] / 1e9), 3
            )

        state["adaptation"] = {
            "measures": adaptation_measures,
            "total_cost": sum(m["cost"] for m in adaptation_measures.values()),
            "lives_protected": int(impacts["economic"]["climate_migration"] * 0.3),
            "economic_benefits": impacts["economic"]["cost_of_inaction"] * 0.4
        }

        return state


class PolicyRecommendationAgent:
    def run(self, state):
        print("🏛️ Generating policy recommendations...")

        mitigation = state["mitigation"]

        recommendations = []

        if mitigation["total_emissions_reduction"] < 0.4:
            recommendations.append({
                "type": "carbon_pricing",
                "urgency": "high",
                "implementation": "immediate",
                "details": "Implement carbon tax at $75-100/ton"
            })

        recommendations.append({
            "type": "renewable_mandate",
            "urgency": "high",
            "implementation": "1-2 years",
            "details": "100% clean electricity by 2035"
        })

        state["policy_recommendations"] = {
            "recommendations": recommendations,
            "total_estimated_impact": round(mitigation["total_emissions_reduction"] * 0.8 + 0.2, 2),
            "political_feasibility": "medium",
            "public_support": 0.65
        }

        return state


class InvestmentAnalysisAgent:
    def __init__(self, investment_horizon=10):
        self.horizon = investment_horizon

    def run(self, state):
        print("💰 Analyzing investment opportunities...")

        mitigation = state["mitigation"]
        adaptation = state["adaptation"]

        investment_opportunities = {
            "renewable_energy": {
                "sector": "energy",
                "growth_rate": 0.12,
                "market_size": 1500,
                "risk_level": "low",
                "climate_alignment": "high"
            },
            "energy_storage": {
                "sector": "technology",
                "growth_rate": 0.25,
                "market_size": 500,
                "risk_level": "medium",
                "climate_alignment": "high"
            },
            "climate_tech": {
                "sector": "technology",
                "growth_rate": 0.30,
                "market_size": 200,
                "risk_level": "high",
                "climate_alignment": "high"
            }
        }

        for opp in investment_opportunities.values():
            opp["estimated_return"] = round(
                (1 + opp["growth_rate"]) ** self.horizon - 1, 2
            )

        state["investment_analysis"] = {
            "opportunities": investment_opportunities,
            "stranded_assets": {
                "fossil_fuel_reserves": {"value_at_risk": 900, "probability": 0.7},
                "coal_power_plants": {"value_at_risk": 150, "probability": 0.9}
            },
            "recommended_allocation": {
                "renewables": 0.4,
                "storage": 0.2,
                "climate_tech": 0.25,
                "adaptation": 0.15
            },
            "total_green_investment_needed": mitigation["total_cost"] + adaptation["total_cost"]
        }

        return state


class SocialEquityAgent:
    def run(self, state):
        print("⚖️ Assessing social equity dimensions...")

        vulnerability_index = {
            "low_income_communities": {
                "exposure": 0.85,
                "sensitivity": 0.90,
                "adaptive_capacity": 0.30
            },
            "indigenous_peoples": {
                "exposure": 0.70,
                "sensitivity": 0.95,
                "adaptive_capacity": 0.40
            }
        }

        state["social_equity"] = {
            "vulnerability_assessment": vulnerability_index,
            "climate_justice_score": self._calculate_justice_score(vulnerability_index),
            "recommended_actions": [
                "40% of climate funds to vulnerable communities",
                "Worker retraining programs"
            ]
        }

        return state

    def _calculate_justice_score(self, vulnerability):
        avg_exposure = np.mean([v["exposure"] for v in vulnerability.values()])
        avg_capacity = np.mean([v["adaptive_capacity"] for v in vulnerability.values()])
        justice_score = 100 * (1 - (avg_exposure * (1 - avg_capacity)))
        return round(justice_score, 1)


class MonitoringVerificationAgent:
    def run(self, state):
        print("📊 Setting up monitoring and verification...")

        kpis = {
            "emissions_tracking": {
                "metric": "CO2_equivalent",
                "target": "45% reduction by 2030",
                "current": "2% reduction",
                "on_track": False
            },
            "renewable_deployment": {
                "metric": "renewable_capacity",
                "target": "triple by 2030",
                "current": "15% increase",
                "on_track": True
            }
        }

        state["monitoring"] = {
            "kpis": kpis,
            "transparency_score": 75,
            "reporting_frequency": "real-time"
        }

        return state


class ScenarioAnalysisAgent:
    def run(self, state):
        print("🎯 Running scenario analysis...")

        scenarios = {
            "best_case": {
                "description": "Aggressive action, tech breakthrough",
                "temperature_rise": 1.5,
                "probability": 0.15
            },
            "likely_case": {
                "description": "Current policies continued",
                "temperature_rise": 2.8,
                "probability": 0.60
            },
            "worst_case": {
                "description": "High emissions, tipping points",
                "temperature_rise": 4.5,
                "probability": 0.25
            }
        }

        expected_temperature = sum(
            s["temperature_rise"] * s["probability"] for s in scenarios.values()
        )

        state["scenario_analysis"] = {
            "scenarios": scenarios,
            "expected_temperature_rise": round(expected_temperature, 2),
            "resilience_score": self._calculate_resilience(scenarios)
        }

        return state

    def _calculate_resilience(self, scenarios):
        worst_case = scenarios["worst_case"]["temperature_rise"]
        resilience = 100 * (1 / worst_case)
        return min(100, round(resilience, 1))


class CommunicationAgent:
    def run(self, state):
        print("📢 Preparing communication materials...")

        audiences = {
            "public": {
                "key_message": "We have solutions, but need urgent action",
                "urgency_level": "high"
            },
            "policy_makers": {
                "key_message": "Climate action is economically beneficial",
                "urgency_level": "critical"
            }
        }

        executive_summary = f"""
        CLIMATE ACTION REPORT - {datetime.now().strftime('%Y-%m-%d')}

        Current Status:
        - Temperature anomaly: {state['raw_data']['temperature_anomaly']}°C

        Projected Outcomes (SSP2-4.5):
        - Temperature by 2100: {state['projections']['ssp2_4.5']['temperature_2100']}°C

        Required Action:
        - Emissions reduction needed: {state['mitigation']['total_emissions_reduction']*100:.0f}%
        - Investment required: ${state['investment_analysis']['total_green_investment_needed']/1e12:.1f} trillion
        """

        state["communication"] = {
            "audience_strategy": audiences,
            "executive_summary": executive_summary
        }

        return state


# -------------------------
# 2. MONTE CARLO AGENTS (SIMPLIFIED)
# -------------------------

class MonteCarloSimulationAgent:
    def __init__(self, n_simulations=5000):
        self.n_simulations = n_simulations

    def run(self, state):
        print(f"🎲 Running {self.n_simulations:,} Monte Carlo simulations...")

        base_temp = state["raw_data"]["temperature_anomaly"]
        base_sea_level = state["raw_data"]["sea_level_rise"]

        # Run simulations
        results = self._run_monte_carlo(base_temp, base_sea_level)

        state["monte_carlo"] = {
            "n_simulations": self.n_simulations,
            "results": results,
            "risk_metrics": self._calculate_risk_metrics(results)
        }

        return state

    def _run_monte_carlo(self, base_temp, base_sea_level):
        """Simplified Monte Carlo simulation"""

        # Define parameter uncertainties
        climate_sensitivity = np.random.lognormal(np.log(3.0), 0.4, self.n_simulations)
        climate_sensitivity = np.clip(climate_sensitivity, 1.5, 6.0)

        emissions_growth = np.random.normal(0.02, 0.01, self.n_simulations)
        tech_progress = np.random.beta(2, 3, self.n_simulations) * 0.09 + 0.01
        policy_effectiveness = np.random.beta(2, 2, self.n_simulations)

        # Calculate temperature projections
        temp_increase = climate_sensitivity * emissions_growth * (1 - tech_progress * policy_effectiveness) * 76
        temp_2100 = base_temp + temp_increase + np.random.normal(0, 0.15, self.n_simulations)

        # Calculate sea level rise
        slr_sensitivity = np.random.lognormal(np.log(0.05), 0.5, self.n_simulations)
        sea_level_2100 = base_sea_level + slr_sensitivity * temp_2100 * 76

        # Calculate economic losses
        economic_loss = 100e12 * 0.0025 * (temp_2100 ** 2)

        # Calculate tipping point risk
        tipping_risk = np.where(temp_2100 > 3.0, 0.9,
                               np.where(temp_2100 > 2.4, 0.5, 0.1))

        return {
            "temperature_2100": temp_2100,
            "sea_level_2100": sea_level_2100,
            "economic_loss": economic_loss,
            "tipping_point_risk": tipping_risk
        }

    def _calculate_risk_metrics(self, results):
        """Calculate risk metrics"""

        metrics = {}

        for var_name, data in results.items():
            metrics[var_name] = {
                "mean": float(np.mean(data)),
                "median": float(np.median(data)),
                "std": float(np.std(data)),
                "p5": float(np.percentile(data, 5)),
                "p95": float(np.percentile(data, 95)),
                "min": float(np.min(data)),
                "max": float(np.max(data))
            }

            if var_name == "temperature_2100":
                metrics[var_name]["prob_above_1.5C"] = float(np.mean(data > 1.5))
                metrics[var_name]["prob_above_2.0C"] = float(np.mean(data > 2.0))
                metrics[var_name]["prob_above_3.0C"] = float(np.mean(data > 3.0))

            elif var_name == "economic_loss":
                metrics[var_name]["prob_loss_above_10T"] = float(np.mean(data > 10e12))
                metrics[var_name]["prob_loss_above_50T"] = float(np.mean(data > 50e12))

        return metrics


# -------------------------
# 3. SIMPLIFIED BAYESIAN AGENTS
# -------------------------

class BayesianLearningAgent:
    def __init__(self, n_samples=1000):
        self.n_samples = n_samples
        self.evidence_history = []

    def run(self, state):
        print("🧠 Performing Bayesian learning...")

        # Record current evidence
        current_temp = state["raw_data"]["temperature_anomaly"]
        current_co2 = state["raw_data"]["ghg_emissions"]["CO2"]

        self.evidence_history.append({
            "timestamp": datetime.now().isoformat(),
            "temperature": current_temp,
            "co2": current_co2
        })

        # Update beliefs using simplified Bayesian inference
        updated_beliefs = self._update_beliefs(state)

        state["bayesian_learning"] = {
            "evidence_count": len(self.evidence_history),
            "updated_beliefs": updated_beliefs,
            "learning_rate": self._calculate_learning_rate(),
            "last_update": datetime.now().isoformat()
        }

        return state

    def _update_beliefs(self, state):
        """Simplified Bayesian belief updating"""

        # Prior beliefs (based on IPCC)
        prior_ecs_mean = 3.0
        prior_ecs_std = 1.0

        # Get evidence
        if len(self.evidence_history) > 1:
            # Calculate temperature trend
            temps = [e["temperature"] for e in self.evidence_history[-5:]]  # Last 5 observations
            temps = temps if len(temps) > 1 else [temps[0], temps[0] + 0.01]

            # Simple linear trend
            x = np.arange(len(temps))
            slope, intercept = np.polyfit(x, temps, 1)

            # Update belief about climate sensitivity based on observed warming
            # Simplified: more warming suggests higher climate sensitivity
            observed_warming_rate = slope * 10  # Convert to °C/decade

            # Bayesian update formula (simplified)
            # Posterior = (Likelihood * Prior) / Evidence
            likelihood_mean = observed_warming_rate * 3.0  # Simplified relationship
            likelihood_std = 0.5

            # Combine prior and likelihood (weighted average)
            prior_weight = 1 / (prior_ecs_std ** 2)
            likelihood_weight = 1 / (likelihood_std ** 2)

            posterior_mean = (prior_ecs_mean * prior_weight + likelihood_mean * likelihood_weight) / (prior_weight + likelihood_weight)
            posterior_std = 1 / np.sqrt(prior_weight + likelihood_weight)

            # Update projections with Bayesian uncertainty
            projections = state.get("projections", {})
            for scenario in projections:
                base_temp = projections[scenario]["temperature_2100"]
                projections[scenario]["bayesian_uncertainty"] = {
                    "mean": base_temp * (posterior_mean / prior_ecs_mean),
                    "lower": base_temp * ((posterior_mean - 1.96 * posterior_std) / prior_ecs_mean),
                    "upper": base_temp * ((posterior_mean + 1.96 * posterior_std) / prior_ecs_mean)
                }

            state["projections"] = projections

            return {
                "climate_sensitivity": {
                    "posterior_mean": float(posterior_mean),
                    "posterior_std": float(posterior_std),
                    "credible_interval": [
                        float(posterior_mean - 1.96 * posterior_std),
                        float(posterior_mean + 1.96 * posterior_std)
                    ]
                },
                "temperature_trend": {
                    "observed_rate": float(observed_warming_rate),
                    "confidence": "high" if len(self.evidence_history) > 10 else "medium"
                }
            }

        return {
            "climate_sensitivity": {
                "posterior_mean": prior_ecs_mean,
                "posterior_std": prior_ecs_std,
                "credible_interval": [prior_ecs_mean - 1.96 * prior_ecs_std, prior_ecs_mean + 1.96 * prior_ecs_std]
            }
        }

    def _calculate_learning_rate(self):
        """Calculate how fast we're learning"""
        if len(self.evidence_history) < 3:
            return 0.0

        # Simple learning rate based on evidence accumulation
        base_rate = 0.5
        evidence_factor = min(1.0, len(self.evidence_history) / 20)  # Cap at 20 observations
        return base_rate * evidence_factor


class SequentialLearningAgent:
    def __init__(self):
        self.learning_history = []

    def run(self, state):
        print("📈 Tracking sequential learning...")

        # Record learning progress
        learning_entry = {
            "timestamp": datetime.now().isoformat(),
            "temperature": state["raw_data"]["temperature_anomaly"],
            "uncertainty_reduction": self._calculate_uncertainty_reduction(state)
        }

        self.learning_history.append(learning_entry)

        # Calculate learning metrics
        if len(self.learning_history) > 1:
            convergence = self._check_convergence()
            learning_speed = self._calculate_learning_speed()

            state["sequential_learning"] = {
                "history": self.learning_history[-10:],  # Last 10 entries
                "total_updates": len(self.learning_history),
                "convergence_status": convergence,
                "learning_speed": learning_speed,
                "recommended_next_update": self._recommend_next_update()
            }

        return state

    def _calculate_uncertainty_reduction(self, state):
        """Calculate reduction in uncertainty"""
        if "monte_carlo" in state:
            temp_std = state["monte_carlo"]["risk_metrics"]["temperature_2100"]["std"]
            # Normalize uncertainty (lower is better)
            max_expected_std = 1.5  # Maximum expected standard deviation
            uncertainty = min(1.0, temp_std / max_expected_std)
            return 1 - uncertainty
        return 0.0

    def _check_convergence(self):
        """Check if beliefs are converging"""
        if len(self.learning_history) < 5:
            return "early_stage"

        # Check temperature stability
        recent_temps = [entry["temperature"] for entry in self.learning_history[-5:]]
        temp_std = np.std(recent_temps)

        if temp_std < 0.1:
            return "converged"
        elif temp_std < 0.2:
            return "converging"
        else:
            return "diverging"

    def _calculate_learning_speed(self):
        """Calculate speed of learning"""
        if len(self.learning_history) < 3:
            return 0.0

        uncertainties = [entry["uncertainty_reduction"] for entry in self.learning_history[-5:]]
        if len(uncertainties) > 1:
            improvements = np.diff(uncertainties)
            return float(np.mean(improvements)) if len(improvements) > 0 else 0.0
        return 0.0

    def _recommend_next_update(self):
        """Recommend when to update next"""
        base_interval = 30  # days
        learning_speed = self._calculate_learning_speed()

        if learning_speed > 0.1:
            # Learning fast, update more frequently
            return base_interval / 2
        elif learning_speed < 0.01:
            # Learning slow, update less frequently
            return base_interval * 2
        else:
            return base_interval


# -------------------------
# 4. VISUALIZATION AGENT
# -------------------------

class VisualizationAgent:
    def __init__(self, save_plots=True):
        self.save_plots = save_plots

    def run(self, state):
        print("📊 Generating visualizations...")

        # Create basic visualizations
        self._create_summary_plot(state)

        if "monte_carlo" in state:
            self._create_monte_carlo_plots(state)

        if "bayesian_learning" in state:
            self._create_bayesian_plots(state)

        state["visualizations"] = {
            "summary_plot": "climate_summary.png",
            "monte_carlo_plot": "monte_carlo_distributions.png",
            "bayesian_plot": "bayesian_learning.png"
        }

        return state

    def _create_summary_plot(self, state):
        """Create summary visualization"""
        fig, axes = plt.subplots(2, 2, figsize=(12, 10))

        # Current status
        if "raw_data" in state:
            data = state["raw_data"]
            metrics = [
                ("Temperature", data["temperature_anomaly"], "°C", "red"),
                ("CO₂", data["ghg_emissions"]["CO2"], "ppm", "blue"),
                ("Renewable Share", data["renewable_share"], "%", "green"),
                ("Sea Level Rise", data["sea_level_rise"], "mm/yr", "purple")
            ]

            for i, (label, value, unit, color) in enumerate(metrics[:4]):
                ax = axes[i // 2, i % 2]
                ax.bar([label], [value], color=color, alpha=0.7)
                ax.set_ylabel(unit)
                ax.set_title(f"Current {label}")
                ax.text(0, value, f"{value}{unit}", ha='center', va='bottom')

        plt.tight_layout()
        if self.save_plots:
            plt.savefig('climate_summary.png', dpi=150, bbox_inches='tight')
        plt.show()

    def _create_monte_carlo_plots(self, state):
        """Create Monte Carlo visualizations"""
        if "monte_carlo" not in state:
            return

        mc_results = state["monte_carlo"]["results"]

        fig, axes = plt.subplots(2, 2, figsize=(14, 10))

        # Temperature distribution
        ax = axes[0, 0]
        ax.hist(mc_results["temperature_2100"], bins=50, alpha=0.7, color='red', edgecolor='black')
        ax.axvline(x=1.5, color='green', linestyle='--', label='1.5°C limit')
        ax.axvline(x=2.0, color='orange', linestyle='--', label='2.0°C limit')
        ax.set_xlabel('Temperature in 2100 (°C)')
        ax.set_ylabel('Frequency')
        ax.set_title('Monte Carlo: Temperature Distribution')
        ax.legend()
        ax.grid(True, alpha=0.3)

        # Economic loss distribution
        ax = axes[0, 1]
        losses_trillions = mc_results["economic_loss"] / 1e12
        ax.hist(losses_trillions, bins=50, alpha=0.7, color='blue', edgecolor='black')
        ax.axvline(x=10, color='red', linestyle='--', label='$10T threshold')
        ax.set_xlabel('Economic Loss (Trillions USD)')
        ax.set_ylabel('Frequency')
        ax.set_title('Monte Carlo: Economic Loss Distribution')
        ax.legend()
        ax.grid(True, alpha=0.3)

        # Risk probabilities
        ax = axes[1, 0]
        risk_metrics = state["monte_carlo"]["risk_metrics"]["temperature_2100"]
        probabilities = {
            '>1.5°C': risk_metrics.get('prob_above_1.5C', 0),
            '>2.0°C': risk_metrics.get('prob_above_2.0C', 0),
            '>3.0°C': risk_metrics.get('prob_above_3.0C', 0)
        }

        colors = ['orange', 'red', 'darkred']
        bars = ax.bar(probabilities.keys(), probabilities.values(), color=colors)
        ax.set_ylabel('Probability')
        ax.set_title('Temperature Threshold Probabilities')
        ax.set_ylim(0, 1)
        ax.grid(True, alpha=0.3)

        for bar, prob in zip(bars, probabilities.values()):
            ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.02,
                   f'{prob:.1%}', ha='center', va='bottom')

        # Scenario comparison
        ax = axes[1, 1]
        if "projections" in state:
            scenarios = list(state["projections"].keys())
            temps = [state["projections"][s]["temperature_2100"] for s in scenarios]

            bars = ax.bar(scenarios, temps, color=['green', 'yellow', 'orange', 'red'])
            ax.set_ylabel('Temperature in 2100 (°C)')
            ax.set_title('Scenario Comparison')
            ax.grid(True, alpha=0.3)

            for bar, temp in zip(bars, temps):
                ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.1,
                       f'{temp}°C', ha='center', va='bottom')

        plt.tight_layout()
        if self.save_plots:
            plt.savefig('monte_carlo_distributions.png', dpi=150, bbox_inches='tight')
        plt.show()

    def _create_bayesian_plots(self, state):
        """Create Bayesian learning visualizations"""
        if "bayesian_learning" not in state:
            return

        bayesian = state["bayesian_learning"]

        fig, axes = plt.subplots(1, 2, figsize=(12, 5))

        # Climate sensitivity belief updating
        ax = axes[0]
        if "updated_beliefs" in bayesian and "climate_sensitivity" in bayesian["updated_beliefs"]:
            cs = bayesian["updated_beliefs"]["climate_sensitivity"]

            # Prior and posterior
            x = np.linspace(1, 6, 1000)
            prior_mean, prior_std = 3.0, 1.0
            posterior_mean, posterior_std = cs["posterior_mean"], cs["posterior_std"]

            prior_pdf = stats.norm.pdf(x, prior_mean, prior_std)
            posterior_pdf = stats.norm.pdf(x, posterior_mean, posterior_std)

            ax.plot(x, prior_pdf, 'r-', linewidth=2, label='Prior (IPCC)')
            ax.plot(x, posterior_pdf, 'b-', linewidth=2, label='Posterior')
            ax.fill_between(x, 0, posterior_pdf, alpha=0.3, color='blue')
            ax.set_xlabel('Climate Sensitivity (°C)')
            ax.set_ylabel('Probability Density')
            ax.set_title('Bayesian Belief Updating')
            ax.legend()
            ax.grid(True, alpha=0.3)

        # Learning progress
        ax = axes[1]
        if "sequential_learning" in state:
            learning = state["sequential_learning"]
            if "history" in learning and len(learning["history"]) > 1:
                history = learning["history"]
                updates = range(len(history))
                temps = [entry["temperature"] for entry in history]

                ax.plot(updates, temps, 'b-o', linewidth=2)
                ax.fill_between(updates,
                               [t - 0.1 for t in temps],
                               [t + 0.1 for t in temps],
                               alpha=0.3, color='blue')
                ax.set_xlabel('Update Number')
                ax.set_ylabel('Temperature Anomaly (°C)')
                ax.set_title('Sequential Learning Progress')
                ax.grid(True, alpha=0.3)

                # Add convergence indicator
                if learning.get("convergence_status") == "converged":
                    ax.text(0.05, 0.95, '✓ CONVERGED', transform=ax.transAxes,
                           verticalalignment='top', color='green',
                           bbox=dict(boxstyle='round', facecolor='lightgreen', alpha=0.8))

        plt.tight_layout()
        if self.save_plots:
            plt.savefig('bayesian_learning.png', dpi=150, bbox_inches='tight')
        plt.show()


# -------------------------
# 5. ORCHESTRATOR & REPORTER
# -------------------------

class ClimateState(dict):
    """Shared memory for all agents"""
    pass


class ClimateOrchestrator:
    def __init__(self, agents, config=None):
        self.agents = agents
        self.config = config or {
            "region": "global",
            "n_monte_carlo": 5000,
            "enable_visualizations": True
        }

    def execute(self):
        print("="*60)
        print("🌍 CLIMATE ACTION SYSTEM")
        print("="*60)

        state = ClimateState()
        state["config"] = self.config
        state["execution_start"] = datetime.now().isoformat()

        for agent in self.agents:
            agent_name = agent.__class__.__name__
            print(f"\n▶️ {agent_name}")

            try:
                state = agent.run(state)
            except Exception as e:
                print(f"   ⚠️ Error in {agent_name}: {str(e)}")
                continue

        state["execution_end"] = datetime.now().isoformat()

        return state


class ClimateReporter:
    @staticmethod
    def generate_report(final_state):
        print("\n" + "="*60)
        print("📋 CLIMATE ACTION REPORT")
        print("="*60)

        # Current Status
        if "raw_data" in final_state:
            data = final_state["raw_data"]
            print(f"\n📊 CURRENT STATUS:")
            print(f"  • Temperature: {data['temperature_anomaly']}°C above pre-industrial")
            print(f"  • CO₂: {data['ghg_emissions']['CO2']} ppm")
            print(f"  • Renewable Energy: {data['renewable_share']}%")

        # Projections
        if "projections" in final_state:
            proj = final_state["projections"]["ssp2_4.5"]
            print(f"\n📈 PROJECTIONS (Middle Scenario):")
            print(f"  • Temperature 2100: {proj['temperature_2100']}°C")
            print(f"  • Sea Level Rise 2100: {proj['sea_level_rise_2100']} cm")

        # Monte Carlo Results
        if "monte_carlo" in final_state:
            mc = final_state["monte_carlo"]
            risk = mc["risk_metrics"]

            print(f"\n🎲 UNCERTAINTY ANALYSIS:")
            print(f"  • Simulations: {mc['n_simulations']:,}")

            if "temperature_2100" in risk:
                temp = risk["temperature_2100"]
                print(f"  • Temperature in 2100:")
                print(f"     Mean: {temp['mean']:.2f}°C")
                print(f"     90% Range: [{temp['p5']:.2f}°C, {temp['p95']:.2f}°C]")
                print(f"     >2°C Probability: {temp.get('prob_above_2.0C', 0):.1%}")
                print(f"     >3°C Probability: {temp.get('prob_above_3.0C', 0):.1%}")

        # Bayesian Learning
        if "bayesian_learning" in final_state:
            bayes = final_state["bayesian_learning"]
            print(f"\n🧠 BAYESIAN LEARNING:")
            print(f"  • Evidence Processed: {bayes.get('evidence_count', 0)}")
            print(f"  • Learning Rate: {bayes.get('learning_rate', 0):.2f}")

            if "updated_beliefs" in bayes and "climate_sensitivity" in bayes["updated_beliefs"]:
                cs = bayes["updated_beliefs"]["climate_sensitivity"]
                print(f"  • Climate Sensitivity:")
                print(f"     Posterior Mean: {cs['posterior_mean']:.2f}°C")
                print(f"     95% Credible Interval: [{cs['credible_interval'][0]:.2f}°C, {cs['credible_interval'][1]:.2f}°C]")

        # Mitigation Strategies
        if "mitigation" in final_state:
            mit = final_state["mitigation"]
            print(f"\n🌿 MITIGATION STRATEGIES:")
            print(f"  • Total Cost: ${mit['total_cost']/1e12:.2f}T")
            print(f"  • Emissions Reduction: {mit['total_emissions_reduction']*100:.1f}%")
            print(f"  • Paris Agreement: {'✅ Compatible' if mit['paris_agreement_compatible'] else '❌ Not Compatible'}")

        # Investment Analysis
        if "investment_analysis" in final_state:
            inv = final_state["investment_analysis"]
            print(f"\n💰 INVESTMENT OPPORTUNITIES:")
            print(f"  • Green Investment Needed: ${inv['total_green_investment_needed']/1e12:.2f}T")
            print(f"  • Recommended Allocation:")
            for sector, allocation in inv["recommended_allocation"].items():
                print(f"     {sector}: {allocation*100:.0f}%")

        # Policy Recommendations
        if "policy_recommendations" in final_state:
            policies = final_state["policy_recommendations"]
            print(f"\n🏛️ POLICY RECOMMENDATIONS:")
            for rec in policies["recommendations"][:3]:  # Top 3
                print(f"  • {rec['type'].replace('_', ' ').title()}: {rec['details']}")

        # Key Actions
        print(f"\n🚀 KEY ACTIONS REQUIRED:")
        print("  1. Implement carbon pricing immediately")
        print("  2. Accelerate renewable energy deployment 3x")
        print("  3. Invest in climate adaptation measures")
        print("  4. Continue monitoring and Bayesian learning")
        print("  5. Update strategies every 2 years based on new evidence")

        print("\n" + "="*60)
        print(f"Report generated: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")
        print("="*60)


# -------------------------
# 6. MAIN EXECUTION
# -------------------------

if __name__ == "__main__":
    # Initialize all agents
    agents = [
        DataCollectionAgent(region="global"),
        ClimateModelingAgent(model_type="IPCC_AR6"),
        ImpactAssessmentAgent(),
        MonteCarloSimulationAgent(n_simulations=3000),
        BayesianLearningAgent(n_samples=1000),
        SequentialLearningAgent(),
        MitigationStrategyAgent(budget_constraint=2e12),
        AdaptationPlanningAgent(),
        InvestmentAnalysisAgent(investment_horizon=10),
        PolicyRecommendationAgent(),
        SocialEquityAgent(),
        MonitoringVerificationAgent(),
        ScenarioAnalysisAgent(),
        VisualizationAgent(save_plots=True),
        CommunicationAgent()
    ]

    # Configuration
    config = {
        "region": "global",
        "n_monte_carlo": 3000,
        "enable_visualizations": True,
        "learning_mode": "adaptive"
    }

    # Execute system
    print("🚀 Starting Climate Action System...")
    orchestrator = ClimateOrchestrator(agents, config)
    final_state = orchestrator.execute()

    # Generate report
    ClimateReporter.generate_report(final_state)

    # Save results
    def simplify_for_json(obj):
        if isinstance(obj, np.ndarray):
            return obj.tolist()
        elif isinstance(obj, np.generic):
            return obj.item()
        elif isinstance(obj, dict):
            return {k: simplify_for_json(v) for k, v in obj.items()}
        elif isinstance(obj, list):
            return [simplify_for_json(v) for v in obj]
        else:
            return obj

    with open("climate_analysis_results.json", "w") as f:
        json.dump(simplify_for_json(final_state), f, indent=2, default=str)

    print("\n✅ Analysis complete! Results saved to 'climate_analysis_results.json'")

    # Optional: Create summary CSV
    try:
        summary_data = []

        # Extract key metrics
        if "raw_data" in final_state:
            summary_data.append(("Current Temperature", f"{final_state['raw_data']['temperature_anomaly']}°C"))
            summary_data.append(("Current CO2", f"{final_state['raw_data']['ghg_emissions']['CO2']} ppm"))

        if "projections" in final_state:
            proj = final_state["projections"]["ssp2_4.5"]
            summary_data.append(("Projected Temperature 2100", f"{proj['temperature_2100']}°C"))

        if "monte_carlo" in final_state:
            risk = final_state["monte_carlo"]["risk_metrics"]
            if "temperature_2100" in risk:
                temp = risk["temperature_2100"]
                summary_data.append(("Probability >2°C", f"{temp.get('prob_above_2.0C', 0):.1%}"))
                summary_data.append(("Probability >3°C", f"{temp.get('prob_above_3.0C', 0):.1%}"))

        if "mitigation" in final_state:
            mit = final_state["mitigation"]
            summary_data.append(("Required Investment", f"${mit['total_cost']/1e12:.2f}T"))
            summary_data.append(("Paris Agreement Compatible", "Yes" if mit['paris_agreement_compatible'] else "No"))

        # Save to CSV
        df = pd.DataFrame(summary_data, columns=["Metric", "Value"])
        df.to_csv("climate_summary.csv", index=False)
        print("📊 Summary saved to 'climate_summary.csv'")

    except Exception as e:
        print(f"Note: Could not create CSV summary: {e}")
```
