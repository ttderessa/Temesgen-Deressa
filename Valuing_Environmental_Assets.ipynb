```python
"""
Willingness to Pay (WTP) using Object-Oriented Programming

Author: Temesgen Deressa
Date: March 31, 2026

Description:
This script models an individual's willingness to pay for environmental
assets
"""


class EnvironmentalAsset:
    def __init__(self, name, area_hectares):
        self.name = name
        self.area = area_hectares

    def calculate_tev(self, unit_use_value, unit_non_use_value):
        """
        Calculates Total Economic Value (TEV).
        TEV = (Use Value + Non-Use Value) * Area
        """
        use_val = unit_use_value * self.area
        non_use_val = unit_non_use_value * self.area
        total_value = use_val + non_use_val

        return {
            "Asset": self.name,
            "Use Value": use_val,
            "Non-Use Value": non_use_val,
            "Total Economic Value": total_value
        }

class ValuationStudy:
    def __init__(self, asset, method_name):
        self.asset = asset
        self.method = method_name

    def run_contingent_valuation(self, mean_wtp, population_size):
        """
        Estimates value based on Mean Willingness to Pay (WTP)
        from a survey sample applied to a broader population.
        """
        estimated_value = mean_wtp * population_size
        print(f"--- {self.method} for {self.asset.name} ---")
        print(f"Mean WTP per person: ${mean_wtp}")
        print(f"Target Population: {population_size}")
        print(f"Aggregate Estimated Value: ${estimated_value:,.2f}\n")
        return estimated_value

# --- Example Usage ---

# 1. Define the Environmental Asset (e.g., a local Wetland)
wetland = EnvironmentalAsset("Chesapeake Mangrove Buffer", area_hectares=500)

# 2. Calculate TEV based on per-hectare ecosystem service estimates
# Assume $200/ha for timber/recreation (Use) and $500/ha for carbon/biodiversity (Non-Use)
values = wetland.calculate_tev(unit_use_value=200, unit_non_use_value=500)

print(f"Asset Analysis: {values['Asset']}")
print(f"Total Economic Value: ${values['Total Economic Value']:,.2f}\n")

# 3. Conduct a Contingent Valuation Study (Survey-based WTP)
study = ValuationStudy(wetland, "Contingent Valuation Method")
study.run_contingent_valuation(mean_wtp=45, population_size=100000)
```
