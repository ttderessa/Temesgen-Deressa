```python
# =============================================================================
# CREDIT RISK ASSESSMENT WITH AGENTIC AI - COMPLETE ADVANCED IMPLEMENTATION
# =============================================================================

import torch
import torch.nn as nn
import torch.optim as optim
import torch.distributions as distributions
from torch.utils.data import DataLoader, TensorDataset
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, QuantileTransformer
from sklearn.model_selection import train_test_split, StratifiedKFold
from sklearn.metrics import (accuracy_score, classification_report,
                           confusion_matrix, roc_auc_score, roc_curve,
                           precision_recall_curve, average_precision_score)
from sklearn.ensemble import IsolationForest
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

# Set random seeds for reproducibility
torch.manual_seed(42)
np.random.seed(42)

# =============================================================================
# 1. ADVANCED DATA GENERATION WITH ECONOMIC CYCLES
# =============================================================================

class AdvancedCreditDataGenerator:
    """Generate sophisticated credit data with economic cycles and regime changes"""

    def __init__(self, n_samples=10000):
        self.n_samples = n_samples
        self.economic_cycles = self._generate_economic_cycles()

    def _generate_economic_cycles(self):
        """Generate realistic economic cycles"""
        time_points = np.linspace(0, 4*np.pi, self.n_samples)
        # Combine multiple economic cycles
        cycle1 = 0.3 * np.sin(time_points)  # Short-term business cycle
        cycle2 = 0.1 * np.sin(0.3 * time_points)  # Medium-term cycle
        cycle3 = 0.05 * np.sin(0.1 * time_points)  # Long-term cycle
        trend = 0.02 * np.arange(self.n_samples) / self.n_samples  # Economic trend

        return cycle1 + cycle2 + cycle3 + trend

    def generate_advanced_credit_data(self):
        """Generate advanced credit data with temporal dependencies"""

        np.random.seed(42)

        # Core financial features with economic cycle influence
        base_age = np.random.normal(45, 12, self.n_samples)
        age = np.clip(base_age + 5 * self.economic_cycles, 18, 80)

        # Income with economic cycle impact
        economic_impact = 0.2 * self.economic_cycles
        base_income = np.random.lognormal(10.8, 0.6, self.n_samples)
        income = base_income * (1 + economic_impact)

        # Credit score with market conditions
        market_impact = 0.15 * self.economic_cycles
        base_credit_score = np.random.normal(680, 80, self.n_samples)
        credit_score = np.clip(base_credit_score + 30 * market_impact, 300, 850)

        # Debt features sensitive to economic conditions
        debt_sensitivity = 0.25 * (1 + self.economic_cycles)
        debt_to_income = np.random.beta(2, 5, self.n_samples) * 1.8 * debt_sensitivity

        # Employment stability
        employment_risk = 0.1 * (1 - self.economic_cycles)  # Higher risk in downturns
        years_employed = np.random.gamma(3, 2, self.n_samples) * (1 - employment_risk)

        # Credit utilization with behavioral patterns
        utilization_trend = 0.3 * self.economic_cycles
        credit_utilization = np.random.beta(2, 3, self.n_samples) * (1 + utilization_trend)

        # Additional advanced features
        num_credit_lines = np.random.poisson(8, self.n_samples)
        recent_inquiries = np.random.poisson(1.2, self.n_samples)
        savings_balance = np.random.lognormal(8, 1.2, self.n_samples)
        investment_assets = np.random.lognormal(9, 1.5, self.n_samples)
        payment_history_score = np.random.beta(8, 2, self.n_samples)  # 0-1 score

        # Macro-economic indicators
        unemployment_rate = 0.05 + 0.03 * self.economic_cycles
        gdp_growth = 0.02 + 0.015 * self.economic_cycles

        # Create sophisticated default probability with regime changes
        default_log_odds = self._calculate_advanced_default_probability(
            age, income, credit_score, debt_to_income, years_employed,
            credit_utilization, num_credit_lines, recent_inquiries,
            savings_balance, investment_assets, payment_history_score,
            unemployment_rate, gdp_growth
        )

        # Convert to probabilities and generate labels with economic timing
        default_probability = 1 / (1 + np.exp(-default_log_odds))

        # Introduce temporal clustering of defaults
        temporal_cluster = np.convolve(default_probability > 0.6,
                                     np.ones(50)/50, mode='same') > 0.1
        default_probability[temporal_cluster] *= 1.3

        default_flag = (default_probability > np.random.uniform(0.3, 0.7, self.n_samples)).astype(int)

        # Create comprehensive DataFrame
        data = pd.DataFrame({
            'age': age,
            'income': income,
            'credit_score': credit_score,
            'debt_to_income': debt_to_income,
            'years_employed': years_employed,
            'credit_utilization': credit_utilization,
            'num_credit_lines': num_credit_lines,
            'recent_inquiries': recent_inquiries,
            'savings_balance': savings_balance,
            'investment_assets': investment_assets,
            'payment_history_score': payment_history_score,
            'unemployment_rate': unemployment_rate,
            'gdp_growth': gdp_growth,
            'economic_cycle': self.economic_cycles,
            'default_probability': default_probability,
            'default_flag': default_flag
        })

        # Remove extreme outliers
        data = data[
            (data['income'] > 15000) & (data['income'] < 300000) &
            (data['age'] >= 18) & (data['age'] <= 80) &
            (data['debt_to_income'] <= 2.0)
        ]

        print(f"Generated {len(data)} advanced samples")
        print(f"Default rate: {data['default_flag'].mean():.3f}")
        print(f"Economic correlation with defaults: {data['economic_cycle'].corr(data['default_flag']):.3f}")

        return data

    def _calculate_advanced_default_probability(self, *features):
        """Calculate sophisticated default probability with interactions"""

        (age, income, credit_score, debt_to_income, years_employed,
         credit_utilization, num_credit_lines, recent_inquiries,
         savings_balance, investment_assets, payment_history_score,
         unemployment_rate, gdp_growth) = features

        # Base linear effects
        base_log_odds = (
            -5.0
            + 0.015 * age
            - 0.000012 * income
            - 0.025 * credit_score
            + 6.0 * debt_to_income
            - 0.4 * years_employed
            + 3.5 * credit_utilization
            + 0.08 * num_credit_lines
            + 0.3 * recent_inquiries
            - 0.000002 * savings_balance
            - 0.000001 * investment_assets
            - 2.5 * payment_history_score
            + 8.0 * unemployment_rate
            - 15.0 * gdp_growth
        )

        # Complex interaction terms
        interactions = (
            + 0.002 * age * debt_to_income  # Age-debt interaction
            - 0.0000001 * income * credit_utilization  # Income-utilization
            + 0.0001 * credit_score * savings_balance  # Credit score-savings
            + 0.5 * debt_to_income * unemployment_rate  # Debt-unemployment
            - 0.3 * years_employed * payment_history_score  # Employment-payment history
            + 0.1 * num_credit_lines * recent_inquiries  # Credit activity
        )

        # Non-linear transformations
        non_linear = (
            + 0.5 * np.log(income + 1)  # Log income effect
            - 0.8 * np.sqrt(credit_score)  # Square root credit score
            + 2.0 * debt_to_income ** 2  # Quadratic debt effect
            - 0.2 * np.exp(0.1 * years_employed)  # Exponential employment
        )

        # Economic regime effects
        regime_effects = (
            + 1.5 * (debt_to_income > 0.5) * (unemployment_rate > 0.06)  # High debt in high unemployment
            - 2.0 * (savings_balance > 50000) * (gdp_growth > 0.02)  # High savings in growth
            + 1.0 * (credit_utilization > 0.8) * (gdp_growth < 0)  # High utilization in recession
        )

        # Add noise with heteroscedasticity (higher uncertainty for riskier applicants)
        risk_score = base_log_odds + interactions + non_linear + regime_effects
        noise_std = 0.5 + 0.3 * torch.sigmoid(torch.tensor(risk_score / 10)).numpy()
        noise = np.random.normal(0, noise_std)

        return risk_score + noise

# Generate advanced data
print("Generating advanced credit data with economic cycles...")
advanced_data_gen = AdvancedCreditDataGenerator(n_samples=10000)
advanced_credit_data = advanced_data_gen.generate_advanced_credit_data()

# Display advanced analysis
print("\nAdvanced Dataset Summary:")
print(advanced_credit_data.describe())

# Plot economic cycle impact
plt.figure(figsize=(15, 10))

plt.subplot(2, 3, 1)
plt.scatter(advanced_credit_data['economic_cycle'],
           advanced_credit_data['default_probability'], alpha=0.5)
plt.xlabel('Economic Cycle')
plt.ylabel('Default Probability')
plt.title('Economic Cycle vs Default Probability')

plt.subplot(2, 3, 2)
sns.kdeplot(data=advanced_credit_data, x='income', hue='default_flag', common_norm=False)
plt.title('Income Distribution by Default Status')

plt.subplot(2, 3, 3)
correlation_matrix = advanced_credit_data.corr()
sns.heatmap(correlation_matrix[['default_flag']].sort_values('default_flag', ascending=False),
            annot=True, cmap='coolwarm', center=0)
plt.title('Feature Correlations with Default')

plt.subplot(2, 3, 4)
defaults_over_time = advanced_credit_data['default_flag'].rolling(window=100, min_periods=1).mean()
plt.plot(defaults_over_time)
plt.xlabel('Time')
plt.ylabel('Default Rate (Rolling Mean)')
plt.title('Temporal Pattern of Defaults')

plt.subplot(2, 3, 5)
# Create risk groups without pandas Interval objects
risk_bins = 5
risk_labels = [f'Group {i+1}' for i in range(risk_bins)]
advanced_credit_data['risk_group'] = pd.cut(advanced_credit_data['default_probability'],
                                          bins=risk_bins, labels=risk_labels)
grouped_data = advanced_credit_data.groupby('risk_group').agg({
    'default_flag': 'mean',
    'income': 'mean',
    'credit_score': 'mean'
}).reset_index()

plt.plot(range(len(grouped_data)), grouped_data['default_flag'], 'o-')
plt.xlabel('Predicted Default Probability Group')
plt.ylabel('Actual Default Rate')
plt.title('Model Calibration')
plt.xticks(range(len(grouped_data)), grouped_data['risk_group'])

plt.subplot(2, 3, 6)
feature_importance = np.abs(correlation_matrix['default_flag']).sort_values(ascending=False)[1:11]
feature_importance.plot(kind='barh')
plt.title('Feature Importance (Absolute Correlation)')

plt.tight_layout()
plt.show()

# =============================================================================
# 2. ADVANCED AGENTIC VAE ARCHITECTURE
# =============================================================================

class MultiHeadAttention(nn.Module):
    """Multi-head attention for capturing feature interactions"""

    def __init__(self, embed_dim, num_heads, dropout=0.1):
        super().__init__()
        self.embed_dim = embed_dim
        self.num_heads = num_heads
        self.head_dim = embed_dim // num_heads

        self.qkv = nn.Linear(embed_dim, embed_dim * 3)
        self.proj = nn.Linear(embed_dim, embed_dim)
        self.dropout = nn.Dropout(dropout)

    def forward(self, x):
        batch_size, seq_len, embed_dim = x.shape

        # Linear projection to Q, K, V
        qkv = self.qkv(x).reshape(batch_size, seq_len, 3, self.num_heads, self.head_dim)
        qkv = qkv.permute(2, 0, 3, 1, 4)  # [3, batch, heads, seq_len, head_dim]
        q, k, v = qkv[0], qkv[1], qkv[2]

        # Scaled dot-product attention
        attn_weights = torch.matmul(q, k.transpose(-2, -1)) / (self.head_dim ** 0.5)
        attn_weights = torch.softmax(attn_weights, dim=-1)
        attn_weights = self.dropout(attn_weights)

        # Apply attention to values
        attn_output = torch.matmul(attn_weights, v)
        attn_output = attn_output.transpose(1, 2).reshape(batch_size, seq_len, embed_dim)

        return self.proj(attn_output)

class AgenticCreditVAE(nn.Module):
    """
    Advanced VAE with agentic capabilities for credit risk assessment
    Features: Multi-head attention, uncertainty quantification, adversarial training
    """

    def __init__(self, input_dim, hidden_dim=256, latent_dim=32, num_heads=4,
                 num_classes=2, dropout=0.2):
        super(AgenticCreditVAE, self).__init__()

        self.input_dim = input_dim
        self.latent_dim = latent_dim
        self.num_heads = num_heads

        # Feature embedding with attention
        self.feature_embedding = nn.Linear(input_dim, hidden_dim)
        self.attention = MultiHeadAttention(hidden_dim, num_heads, dropout)
        self.layer_norm1 = nn.LayerNorm(hidden_dim)
        self.layer_norm2 = nn.LayerNorm(hidden_dim)

        # Advanced encoder with residual connections
        self.encoder_blocks = nn.ModuleList([
            nn.Sequential(
                nn.Linear(hidden_dim, hidden_dim),
                nn.GELU(),
                nn.Dropout(dropout),
                nn.Linear(hidden_dim, hidden_dim),
            ) for _ in range(2)  # Reduced for efficiency
        ])

        # Latent space parameters
        self.fc_mu = nn.Linear(hidden_dim, latent_dim)
        self.fc_logvar = nn.Linear(hidden_dim, latent_dim)

        # Uncertainty estimation
        self.aleatoric_net = nn.Sequential(
            nn.Linear(latent_dim, 32),
            nn.GELU(),
            nn.Linear(32, num_classes)
        )

        # Advanced decoder
        self.decoder = nn.Sequential(
            nn.Linear(latent_dim, hidden_dim),
            nn.GELU(),
            nn.Dropout(dropout),
            nn.Linear(hidden_dim, hidden_dim),
            nn.GELU(),
            nn.Linear(hidden_dim, input_dim),
            nn.Tanh()
        )

        # Multi-task risk classifier with uncertainty
        self.risk_classifier = nn.Sequential(
            nn.Linear(latent_dim, 128),
            nn.GELU(),
            nn.Dropout(dropout),
            nn.Linear(128, 64),
            nn.GELU(),
            nn.Linear(64, num_classes)
        )

        # Economic sensitivity network
        self.economic_sensitivity = nn.Sequential(
            nn.Linear(latent_dim + 2, 32),  # +2 for economic indicators
            nn.GELU(),
            nn.Linear(32, 1),
            nn.Sigmoid()
        )

        # Adversarial component for robustness
        self.adversarial_net = nn.Sequential(
            nn.Linear(latent_dim, 32),
            nn.GELU(),
            nn.Linear(32, 1),
            nn.Sigmoid()
        )

        self.dropout = nn.Dropout(dropout)

    def encode(self, x):
        """Advanced encoding with attention"""
        # Feature embedding and attention
        h = self.feature_embedding(x).unsqueeze(1)  # Add sequence dimension
        h_attn = self.attention(h)
        h = self.layer_norm1(h + h_attn)
        h = h.squeeze(1)

        # Residual blocks
        for block in self.encoder_blocks:
            residual = h
            h = block(h)
            h = self.layer_norm2(h + residual)

        return self.fc_mu(h), self.fc_logvar(h)

    def reparameterize(self, mu, logvar):
        """Reparameterization with temperature sampling"""
        std = torch.exp(0.5 * logvar)
        eps = torch.randn_like(std)

        # Temperature sampling for exploration
        temperature = 0.8 + 0.4 * torch.sigmoid(-logvar.mean())  # Adaptive temperature
        return mu + temperature * eps * std

    def decode(self, z):
        return self.decoder(z)

    def forward(self, x, economic_indicators=None):
        """Forward pass with multiple outputs"""
        mu, logvar = self.encode(x)
        z = self.reparameterize(mu, logvar)
        reconstruction = self.decode(z)

        # Risk classification
        risk_logits = self.risk_classifier(z)

        # Aleatoric uncertainty
        aleatoric_logits = self.aleatoric_net(z)

        # Economic sensitivity if indicators provided
        economic_sensitivity = None
        if economic_indicators is not None:
            economic_input = torch.cat([z, economic_indicators], dim=1)
            economic_sensitivity = self.economic_sensitivity(economic_input)

        # Adversarial output
        adversarial_out = self.adversarial_net(z)

        return {
            'reconstruction': reconstruction,
            'mu': mu,
            'logvar': logvar,
            'z': z,
            'risk_logits': risk_logits,
            'aleatoric_logits': aleatoric_logits,
            'economic_sensitivity': economic_sensitivity,
            'adversarial_out': adversarial_out
        }

    def generate_counterfactuals(self, x, n_variations=3, variation_strength=0.3):
        """Generate counterfactual explanations"""
        with torch.no_grad():
            mu, logvar = self.encode(x)
            counterfactuals = []

            for i in range(n_variations):
                # Vary along different latent dimensions
                z_variation = mu.clone()
                variation_dim = i % self.latent_dim
                z_variation[:, variation_dim] += variation_strength * torch.randn_like(z_variation[:, variation_dim])

                counterfactual = self.decode(z_variation)
                counterfactuals.append(counterfactual)

            return torch.stack(counterfactuals)

    def estimate_uncertainty(self, x, n_samples=20):
        """Estimate epistemic and aleatoric uncertainty"""
        with torch.no_grad():
            risk_probs = []

            for _ in range(n_samples):
                # Enable dropout for uncertainty estimation
                self.train()
                outputs = self.forward(x)
                risk_prob = torch.softmax(outputs['risk_logits'], dim=1)
                risk_probs.append(risk_prob)

            risk_probs = torch.stack(risk_probs)
            epistemic_uncertainty = risk_probs.std(dim=0).mean(dim=1)

            # Aleatoric uncertainty from the network
            self.eval()
            outputs = self.forward(x)
            aleatoric_uncertainty = torch.softmax(outputs['aleatoric_logits'], dim=1).std(dim=1)

            return {
                'epistemic': epistemic_uncertainty,
                'aleatoric': aleatoric_uncertainty,
                'total': epistemic_uncertainty + aleatoric_uncertainty
            }

# =============================================================================
# 3. ADVANCED TRAINING FRAMEWORK WITH MULTIPLE LOSSES
# =============================================================================

class AgenticTrainer:
    """Advanced trainer with multiple loss functions and adversarial training"""

    def __init__(self, model, learning_rate=1e-3, alpha=0.1, beta=0.01, gamma=0.05):
        self.model = model
        self.optimizer = optim.AdamW(model.parameters(), lr=learning_rate, weight_decay=1e-4)
        self.scheduler = optim.lr_scheduler.CosineAnnealingWarmRestarts(
            self.optimizer, T_0=10, T_mult=2
        )

        # Loss weights
        self.alpha = alpha  # KL divergence weight
        self.beta = beta    # Adversarial weight
        self.gamma = gamma  # Economic sensitivity weight

        # Loss tracking
        self.loss_history = {
            'total': [], 'reconstruction': [], 'kl': [], 'classification': [],
            'adversarial': [], 'economic': [], 'aleatoric': []
        }

    def vae_loss(self, reconstruction, x, mu, logvar):
        """Enhanced VAE loss with beta-VAE"""
        reconstruction_loss = nn.functional.mse_loss(reconstruction, x, reduction='mean')
        kl_divergence = -0.5 * torch.mean(1 + logvar - mu.pow(2) - logvar.exp())

        # Beta-VAE with adaptive weighting
        beta = self.alpha * (1 + torch.sigmoid(torch.tensor(kl_divergence / 1000)))
        return reconstruction_loss + beta * kl_divergence

    def adversarial_loss(self, adversarial_out, labels):
        """Adversarial loss for robustness"""
        # Encourage model to be invariant to sensitive variations
        return torch.mean(adversarial_out)

    def economic_sensitivity_loss(self, economic_sensitivity, economic_indicators, labels):
        """Loss for economic sensitivity calibration"""
        if economic_sensitivity is None:
            return torch.tensor(0.0)

        # Penalize miscalibration of economic sensitivity
        expected_sensitivity = labels.float() * economic_indicators[:, 0]  # unemployment impact
        return nn.functional.mse_loss(economic_sensitivity.squeeze(), expected_sensitivity)

    def aleatoric_loss(self, aleatoric_logits, risk_logits, labels):
        """Loss for aleatoric uncertainty calibration"""
        risk_probs = torch.softmax(risk_logits, dim=1)
        aleatoric_probs = torch.softmax(aleatoric_logits, dim=1)

        # Encourage higher uncertainty for misclassified samples
        correct_predictions = (risk_probs.argmax(dim=1) == labels).float()
        uncertainty_penalty = aleatoric_probs.std(dim=1)

        return torch.mean(uncertainty_penalty * correct_predictions)

    def train_epoch(self, data_loader, device, epoch):
        """Advanced training with multiple losses"""
        self.model.train()
        total_loss = 0

        for batch_idx, (data, targets, economic_indicators) in enumerate(data_loader):
            data, targets, economic_indicators = (data.to(device),
                                                targets.to(device),
                                                economic_indicators.to(device))

            self.optimizer.zero_grad()

            # Forward pass with economic indicators
            outputs = self.model(data, economic_indicators)

            # Calculate individual losses
            reconstruction_loss = self.vae_loss(
                outputs['reconstruction'], data,
                outputs['mu'], outputs['logvar']
            )

            classification_loss = nn.functional.cross_entropy(
                outputs['risk_logits'], targets
            )

            adversarial_loss = self.adversarial_loss(
                outputs['adversarial_out'], targets
            )

            economic_loss = self.economic_sensitivity_loss(
                outputs['economic_sensitivity'], economic_indicators, targets
            )

            aleatoric_loss = self.aleatoric_loss(
                outputs['aleatoric_logits'], outputs['risk_logits'], targets
            )

            # Combined loss
            loss = (reconstruction_loss + classification_loss +
                   self.beta * adversarial_loss +
                   self.gamma * economic_loss +
                   aleatoric_loss)

            # Backward pass with gradient clipping
            loss.backward()
            torch.nn.utils.clip_grad_norm_(self.model.parameters(), max_norm=1.0)
            self.optimizer.step()

            total_loss += loss.item()

            # Update loss history
            self.loss_history['total'].append(loss.item())
            self.loss_history['reconstruction'].append(reconstruction_loss.item())
            self.loss_history['classification'].append(classification_loss.item())
            self.loss_history['adversarial'].append(adversarial_loss.item())
            self.loss_history['economic'].append(economic_loss.item())
            self.loss_history['aleatoric'].append(aleatoric_loss.item())

            if batch_idx % 50 == 0:
                print(f'Epoch {epoch}, Batch {batch_idx}: Loss = {loss.item():.4f}')

        self.scheduler.step()
        return total_loss / len(data_loader)

    def evaluate(self, data_loader, device):
        """Comprehensive evaluation"""
        self.model.eval()
        total_loss = 0
        all_predictions = []
        all_targets = []
        all_uncertainties = []

        with torch.no_grad():
            for data, targets, economic_indicators in data_loader:
                data, targets = data.to(device), targets.to(device)

                outputs = self.model(data)
                loss = nn.functional.cross_entropy(outputs['risk_logits'], targets)
                total_loss += loss.item()

                # Predictions and uncertainties
                risk_probs = torch.softmax(outputs['risk_logits'], dim=1)
                predictions = risk_probs.argmax(dim=1)

                all_predictions.extend(predictions.cpu().numpy())
                all_targets.extend(targets.cpu().numpy())

                # Estimate uncertainty
                uncertainty = self.model.estimate_uncertainty(data)
                all_uncertainties.append(uncertainty['total'].mean().item())

        accuracy = accuracy_score(all_targets, all_predictions)
        avg_uncertainty = np.mean(all_uncertainties)

        return total_loss / len(data_loader), accuracy, avg_uncertainty

# =============================================================================
# 4. ADVANCED DATA PREPARATION
# =============================================================================

def prepare_advanced_training_data(credit_data, test_size=0.15, validation_size=0.15):
    """Prepare advanced training data with economic indicators"""

    # Select features - expanded set
    feature_columns = [
        'age', 'income', 'credit_score', 'debt_to_income',
        'years_employed', 'credit_utilization', 'num_credit_lines',
        'recent_inquiries', 'savings_balance', 'investment_assets',
        'payment_history_score'
    ]

    # Economic indicators
    economic_columns = ['unemployment_rate', 'gdp_growth']

    X = credit_data[feature_columns].values
    E = credit_data[economic_columns].values  # Economic indicators
    y = credit_data['default_flag'].values

    # Advanced preprocessing
    feature_scaler = StandardScaler()
    economic_scaler = StandardScaler()

    X_scaled = feature_scaler.fit_transform(X)
    E_scaled = economic_scaler.fit_transform(E)

    # Advanced data splitting with stratification
    X_temp, X_test, E_temp, E_test, y_temp, y_test = train_test_split(
        X_scaled, E_scaled, y, test_size=test_size,
        random_state=42, stratify=y
    )

    val_size = validation_size / (1 - test_size)
    X_train, X_val, E_train, E_val, y_train, y_val = train_test_split(
        X_temp, E_temp, y_temp, test_size=val_size,
        random_state=42, stratify=y_temp
    )

    # Convert to PyTorch tensors
    X_train_tensor = torch.FloatTensor(X_train)
    E_train_tensor = torch.FloatTensor(E_train)
    y_train_tensor = torch.LongTensor(y_train)

    X_val_tensor = torch.FloatTensor(X_val)
    E_val_tensor = torch.FloatTensor(E_val)
    y_val_tensor = torch.LongTensor(y_val)

    X_test_tensor = torch.FloatTensor(X_test)
    E_test_tensor = torch.FloatTensor(E_test)
    y_test_tensor = torch.LongTensor(y_test)

    print(f"Training samples: {len(X_train)}")
    print(f"Validation samples: {len(X_val)}")
    print(f"Test samples: {len(X_test)}")
    print(f"Default rate - Train: {y_train.mean():.3f}, Test: {y_test.mean():.3f}")

    return (X_train_tensor, E_train_tensor, y_train_tensor,
            X_val_tensor, E_val_tensor, y_val_tensor,
            X_test_tensor, E_test_tensor, y_test_tensor,
            feature_scaler, economic_scaler, feature_columns)

print("Preparing advanced training data...")
(X_train, E_train, y_train, X_val, E_val, y_val,
 X_test, E_test, y_test, feature_scaler, economic_scaler,
 feature_columns) = prepare_advanced_training_data(advanced_credit_data)

# Create advanced data loaders
class AdvancedCreditDataset(torch.utils.data.Dataset):
    def __init__(self, features, economic_indicators, labels):
        self.features = features
        self.economic_indicators = economic_indicators
        self.labels = labels

    def __len__(self):
        return len(self.features)

    def __getitem__(self, idx):
        return (self.features[idx], self.labels[idx],
                self.economic_indicators[idx])

train_dataset = AdvancedCreditDataset(X_train, E_train, y_train)
val_dataset = AdvancedCreditDataset(X_val, E_val, y_val)
test_dataset = AdvancedCreditDataset(X_test, E_test, y_test)

train_loader = DataLoader(train_dataset, batch_size=128, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=128, shuffle=False)
test_loader = DataLoader(test_dataset, batch_size=128, shuffle=False)

# =============================================================================
# 5. ADVANCED MODEL TRAINING
# =============================================================================

def train_agentic_model(train_loader, val_loader, input_dim, epochs=80):
    """Train the advanced agentic model"""

    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    print(f"Using device: {device}")

    # Initialize advanced model
    model = AgenticCreditVAE(
        input_dim=input_dim,
        hidden_dim=256,
        latent_dim=32,
        num_heads=4,
        dropout=0.2
    ).to(device)

    trainer = AgenticTrainer(
        model,
        learning_rate=1e-3,
        alpha=0.05,  # KL weight
        beta=0.01,   # Adversarial weight
        gamma=0.02   # Economic sensitivity weight
    )

    # Training history
    train_losses = []
    val_losses = []
    val_accuracies = []
    uncertainties = []

    best_val_accuracy = 0
    patience = 15
    patience_counter = 0

    print("Starting advanced training...")
    for epoch in range(epochs):
        # Training
        train_loss = trainer.train_epoch(train_loader, device, epoch)

        # Validation
        val_loss, val_accuracy, avg_uncertainty = trainer.evaluate(val_loader, device)

        # Save history
        train_losses.append(train_loss)
        val_losses.append(val_loss)
        val_accuracies.append(val_accuracy)
        uncertainties.append(avg_uncertainty)

        # Early stopping check
        if val_accuracy > best_val_accuracy:
            best_val_accuracy = val_accuracy
            patience_counter = 0
            # Save best model
            torch.save({
                'epoch': epoch,
                'model_state_dict': model.state_dict(),
                'optimizer_state_dict': trainer.optimizer.state_dict(),
                'best_accuracy': best_val_accuracy,
            }, 'best_agentic_model.pth')
        else:
            patience_counter += 1

        if epoch % 10 == 0:
            print(f'Epoch {epoch:3d}: '
                  f'Train Loss = {train_loss:.4f}, '
                  f'Val Loss = {val_loss:.4f}, '
                  f'Val Acc = {val_accuracy:.4f}, '
                  f'Uncertainty = {avg_uncertainty:.4f}')

        if patience_counter >= patience:
            print(f"Early stopping at epoch {epoch}")
            break

    # Load best model
    checkpoint = torch.load('best_agentic_model.pth')
    model.load_state_dict(checkpoint['model_state_dict'])

    print(f"Best validation accuracy: {best_val_accuracy:.4f}")

    return model, trainer, train_losses, val_losses, val_accuracies, uncertainties

print("Training advanced agentic model...")
input_dim = X_train.shape[1]
agentic_model, agentic_trainer, train_losses, val_losses, val_accuracies, uncertainties = train_agentic_model(
    train_loader, val_loader, input_dim, epochs=80
)

# Plot advanced training history
plt.figure(figsize=(20, 5))

plt.subplot(1, 4, 1)
plt.plot(train_losses, label='Training Loss')
plt.plot(val_losses, label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.title('Training and Validation Loss')

plt.subplot(1, 4, 2)
plt.plot(val_accuracies)
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Validation Accuracy')

plt.subplot(1, 4, 3)
plt.plot(uncertainties)
plt.xlabel('Epoch')
plt.ylabel('Uncertainty')
plt.title('Model Uncertainty Over Time')

plt.subplot(1, 4, 4)
# Plot individual loss components (last 100 batches)
for loss_name, loss_values in agentic_trainer.loss_history.items():
    if len(loss_values) > 100:
        plt.plot(loss_values[-100:], label=loss_name, alpha=0.7)
plt.xlabel('Batch')
plt.ylabel('Loss')
plt.legend()
plt.title('Loss Components (Recent)')

plt.tight_layout()
plt.show()

# =============================================================================
# 6. ADVANCED RISK ASSESSMENT WITH EXPLAINABLE AI
# =============================================================================

class AgenticRiskInterpreter:
    """Advanced risk interpretation with explainable AI capabilities"""

    def __init__(self, model, feature_scaler, economic_scaler, feature_columns):
        self.model = model
        self.feature_scaler = feature_scaler
        self.economic_scaler = economic_scaler
        self.feature_columns = feature_columns
        self.device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

    def assess_risk_with_explanations(self, applicant_features, economic_indicators=None):
        """Comprehensive risk assessment with detailed explanations"""

        # Scale features
        scaled_features = self.feature_scaler.transform([applicant_features])
        features_tensor = torch.FloatTensor(scaled_features).to(self.device)

        if economic_indicators is not None:
            scaled_economic = self.economic_scaler.transform([economic_indicators])
            economic_tensor = torch.FloatTensor(scaled_economic).to(self.device)
        else:
            economic_tensor = None

        # Get comprehensive assessment
        self.model.eval()
        with torch.no_grad():
            outputs = self.model(features_tensor, economic_tensor)

            # Risk probability
            risk_proba = torch.softmax(outputs['risk_logits'], dim=1)
            default_prob = risk_proba[0, 1].item()

            # Uncertainty estimates
            uncertainty = self.model.estimate_uncertainty(features_tensor)
            epistemic_uncertainty = uncertainty['epistemic'].item()
            aleatoric_uncertainty = uncertainty['aleatoric'].item()
            total_uncertainty = uncertainty['total'].item()

            # Economic sensitivity
            economic_sensitivity = None
            if outputs['economic_sensitivity'] is not None:
                economic_sensitivity = outputs['economic_sensitivity'][0, 0].item()

            # Generate counterfactuals
            counterfactuals = self.model.generate_counterfactuals(features_tensor, n_variations=3)
            counterfactual_risks = []

            for cf in counterfactuals:
                cf_outputs = self.model(cf)
                cf_risk = torch.softmax(cf_outputs['risk_logits'], dim=1)[0, 1].item()
                counterfactual_risks.append(cf_risk)

        # Risk categorization with uncertainty adjustment
        base_risk_category = self._categorize_risk(default_prob)

        # Adjust for uncertainty
        if total_uncertainty > 0.1:
            risk_category = f"{base_risk_category} (High Uncertainty)"
        elif total_uncertainty > 0.05:
            risk_category = f"{base_risk_category} (Medium Uncertainty)"
        else:
            risk_category = f"{base_risk_category} (Low Uncertainty)"

        # Feature contributions
        feature_contributions = self._calculate_feature_contributions(
            applicant_features, default_prob
        )

        return {
            'default_probability': default_prob,
            'risk_category': risk_category,
            'epistemic_uncertainty': epistemic_uncertainty,
            'aleatoric_uncertainty': aleatoric_uncertainty,
            'total_uncertainty': total_uncertainty,
            'economic_sensitivity': economic_sensitivity,
            'counterfactual_risks': counterfactual_risks,
            'feature_contributions': feature_contributions,
            'recommendation': self._get_recommendation(default_prob, total_uncertainty)
        }

    def _calculate_feature_contributions(self, applicant_features, base_prob):
        """Calculate SHAP-like feature contributions"""
        contributions = {}
        base_scaled = self.feature_scaler.transform([applicant_features])
        base_tensor = torch.FloatTensor(base_scaled).to(self.device)

        for i, feature in enumerate(self.feature_columns):
            # Perturb feature and measure impact
            perturbed_features = base_scaled.copy()
            original_value = perturbed_features[0, i]

            # Positive perturbation
            perturbed_features[0, i] = original_value + 0.1
            pos_tensor = torch.FloatTensor(perturbed_features).to(self.device)
            with torch.no_grad():
                pos_outputs = self.model(pos_tensor)
                pos_prob = torch.softmax(pos_outputs['risk_logits'], dim=1)[0, 1].item()

            # Negative perturbation
            perturbed_features[0, i] = original_value - 0.1
            neg_tensor = torch.FloatTensor(perturbed_features).to(self.device)
            with torch.no_grad():
                neg_outputs = self.model(neg_tensor)
                neg_prob = torch.softmax(neg_outputs['risk_logits'], dim=1)[0, 1].item()

            # Calculate average impact
            pos_impact = pos_prob - base_prob
            neg_impact = base_prob - neg_prob
            avg_impact = (pos_impact + neg_impact) / 2

            contributions[feature] = {
                'impact': avg_impact,
                'direction': 'increases' if avg_impact > 0 else 'decreases',
                'magnitude': abs(avg_impact)
            }

        return dict(sorted(contributions.items(),
                         key=lambda x: abs(x[1]['magnitude']), reverse=True))

    def _categorize_risk(self, prob):
        """Advanced risk categorization"""
        if prob < 0.03:
            return 'Very Low'
        elif prob < 0.08:
            return 'Low'
        elif prob < 0.15:
            return 'Medium-Low'
        elif prob < 0.25:
            return 'Medium'
        elif prob < 0.4:
            return 'Medium-High'
        elif prob < 0.6:
            return 'High'
        else:
            return 'Very High'

    def _get_recommendation(self, prob, uncertainty):
        """Generate sophisticated recommendations"""
        if prob < 0.05:
            base_rec = 'Approve with preferential terms'
        elif prob < 0.15:
            base_rec = 'Approve with standard terms'
        elif prob < 0.3:
            base_rec = 'Approve with enhanced monitoring and conditions'
        elif prob < 0.5:
            base_rec = 'Consider with significant collateral or higher rates'
        else:
            base_rec = 'Reject or require exceptional collateral'

        # Add uncertainty qualification
        if uncertainty > 0.1:
            base_rec += ' (High uncertainty - consider manual review)'
        elif uncertainty > 0.05:
            base_rec += ' (Moderate uncertainty)'

        return base_rec

    def generate_comprehensive_report(self, applicant_features, economic_indicators=None,
                                    applicant_name="Applicant"):
        """Generate comprehensive risk assessment report"""

        assessment = self.assess_risk_with_explanations(applicant_features, economic_indicators)

        print("="*70)
        print(f"ADVANCED CREDIT RISK ASSESSMENT REPORT")
        print(f"Applicant: {applicant_name}")
        print("="*70)

        print("\nAPPLICANT PROFILE:")
        for feature, value in zip(self.feature_columns, applicant_features):
            print(f"  {feature:25}: {value:>10.2f}")

        if economic_indicators is not None:
            print(f"\nECONOMIC CONTEXT:")
            print(f"  Unemployment Rate: {economic_indicators[0]:.3f}")
            print(f"  GDP Growth: {economic_indicators[1]:.3f}")

        print(f"\nRISK ASSESSMENT:")
        print(f"  Default Probability: {assessment['default_probability']:.4f}")
        print(f"  Risk Category: {assessment['risk_category']}")
        print(f"  Total Uncertainty: {assessment['total_uncertainty']:.4f}")
        print(f"  Economic Sensitivity: {assessment['economic_sensitivity']:.4f}"
              if assessment['economic_sensitivity'] is not None else "  Economic Sensitivity: N/A")
        print(f"  Recommendation: {assessment['recommendation']}")

        print(f"\nTOP FEATURE CONTRIBUTIONS:")
        top_features = list(assessment['feature_contributions'].items())[:6]
        for feature, contrib in top_features:
            print(f"  {feature:25}: {contrib['direction']} risk by {abs(contrib['impact']):.4f}")

        print(f"\nCOUNTERFACTUAL ANALYSIS:")
        for i, cf_risk in enumerate(assessment['counterfactual_risks']):
            change = cf_risk - assessment['default_probability']
            direction = "increase" if change > 0 else "decrease"
            print(f"  Scenario {i+1}: {abs(change):.4f} {direction} in default probability")

        print(f"\nUNCERTAINTY BREAKDOWN:")
        print(f"  Epistemic (Model): {assessment['epistemic_uncertainty']:.4f}")
        print(f"  Aleatoric (Data): {assessment['aleatoric_uncertainty']:.4f}")

        print("="*70)

# Initialize advanced interpreter
print("Initializing advanced risk interpreter...")
agentic_interpreter = AgenticRiskInterpreter(
    agentic_model, feature_scaler, economic_scaler, feature_columns
)

# Example comprehensive assessments
print("\nADVANCED CREDIT ASSESSMENTS:\n")

# Low-risk applicant in normal economy
low_risk_applicant = [35, 85000, 780, 0.18, 10, 0.25, 4, 1, 50000, 100000, 0.95]
normal_economy = [0.05, 0.02]
agentic_interpreter.generate_comprehensive_report(
    low_risk_applicant, normal_economy, "Low-Risk Example"
)

# Medium-risk applicant in recession
medium_risk_applicant = [48, 55000, 640, 0.42, 4, 0.65, 9, 3, 15000, 25000, 0.75]
recession_economy = [0.08, -0.01]
agentic_interpreter.generate_comprehensive_report(
    medium_risk_applicant, recession_economy, "Medium-Risk in Recession"
)

# =============================================================================
# 7. COMPREHENSIVE MODEL EVALUATION
# =============================================================================

def comprehensive_agentic_evaluation(model, test_loader, device):
    """Comprehensive evaluation of the agentic model"""

    model.eval()
    all_predictions = []
    all_probabilities = []
    all_targets = []
    all_uncertainties = []

    with torch.no_grad():
        for data, targets, economic_indicators in test_loader:
            data, targets = data.to(device), targets.to(device)

            outputs = model(data)
            risk_proba = torch.softmax(outputs['risk_logits'], dim=1)
            predictions = risk_proba.argmax(dim=1)

            all_predictions.extend(predictions.cpu().numpy())
            all_probabilities.extend(risk_proba[:, 1].cpu().numpy())
            all_targets.extend(targets.cpu().numpy())

            # Uncertainty estimates
            uncertainty = model.estimate_uncertainty(data)
            all_uncertainties.extend(uncertainty['total'].cpu().numpy())

    # Calculate comprehensive metrics
    accuracy = accuracy_score(all_targets, all_predictions)
    auc = roc_auc_score(all_targets, all_probabilities)
    avg_uncertainty = np.mean(all_uncertainties)

    # Precision-Recall curve
    precision, recall, _ = precision_recall_curve(all_targets, all_probabilities)
    avg_precision = average_precision_score(all_targets, all_probabilities)

    # Brier score
    brier_score = np.mean((np.array(all_probabilities) - np.array(all_targets)) ** 2)

    print("COMPREHENSIVE AGENTIC MODEL EVALUATION")
    print("="*50)
    print(f"Test Accuracy: {accuracy:.4f}")
    print(f"Test AUC: {auc:.4f}")
    print(f"Average Precision: {avg_precision:.4f}")
    print(f"Brier Score: {brier_score:.4f}")
    print(f"Average Uncertainty: {avg_uncertainty:.4f}")

    print("\nClassification Report:")
    print(classification_report(all_targets, all_predictions,
                              target_names=['Non-Default', 'Default']))

    # Plot comprehensive evaluation
    plt.figure(figsize=(20, 5))

    # ROC Curve
    plt.subplot(1, 4, 1)
    fpr, tpr, _ = roc_curve(all_targets, all_probabilities)
    plt.plot(fpr, tpr, linewidth=2, label=f'Agentic Model (AUC = {auc:.4f})')
    plt.plot([0, 1], [0, 1], 'k--', label='Random Classifier')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend()
    plt.grid(True, alpha=0.3)

    # Precision-Recall Curve
    plt.subplot(1, 4, 2)
    plt.plot(recall, precision, linewidth=2, label=f'AP = {avg_precision:.4f}')
    plt.xlabel('Recall')
    plt.ylabel('Precision')
    plt.title('Precision-Recall Curve')
    plt.legend()
    plt.grid(True, alpha=0.3)

    # Calibration plot
    plt.subplot(1, 4, 3)
    prob_bins = np.linspace(0, 1, 11)
    bin_centers = (prob_bins[:-1] + prob_bins[1:]) / 2
    empirical_probs = []

    for i in range(len(prob_bins) - 1):
        mask = (np.array(all_probabilities) >= prob_bins[i]) & (np.array(all_probabilities) < prob_bins[i + 1])
        if mask.sum() > 0:
            empirical_prob = np.array(all_targets)[mask].mean()
            empirical_probs.append(empirical_prob)
        else:
            empirical_probs.append(0)

    plt.plot(bin_centers, empirical_probs, 's-', label='Agentic Model')
    plt.plot([0, 1], [0, 1], 'k--', label='Perfect Calibration')
    plt.xlabel('Predicted Probability')
    plt.ylabel('Empirical Probability')
    plt.title('Calibration Plot')
    plt.legend()
    plt.grid(True, alpha=0.3)

    # Uncertainty analysis
    plt.subplot(1, 4, 4)
    correct_predictions = (np.array(all_predictions) == np.array(all_targets))
    plt.boxplot([np.array(all_uncertainties)[correct_predictions],
                np.array(all_uncertainties)[~correct_predictions]],
               labels=['Correct', 'Incorrect'])
    plt.ylabel('Uncertainty')
    plt.title('Uncertainty by Prediction Accuracy')
    plt.grid(True, alpha=0.3)

    plt.tight_layout()
    plt.show()

    # Confusion matrix
    plt.figure(figsize=(8, 6))
    cm = confusion_matrix(all_targets, all_predictions)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['Non-Default', 'Default'],
                yticklabels=['Non-Default', 'Default'])
    plt.title('Confusion Matrix - Agentic Model')
    plt.ylabel('True Label')
    plt.xlabel('Predicted Label')
    plt.show()

    return {
        'accuracy': accuracy,
        'auc': auc,
        'avg_precision': avg_precision,
        'brier_score': brier_score,
        'avg_uncertainty': avg_uncertainty
    }

print("Performing comprehensive agentic model evaluation...")
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
agentic_metrics = comprehensive_agentic_evaluation(agentic_model, test_loader, device)

# =============================================================================
# FINAL SUMMARY
# =============================================================================

print("\n" + "="*80)
print("AGENTIC AI CREDIT RISK ASSESSMENT - COMPLETE IMPLEMENTATION")
print("="*80)

# Performance summary
print("\nPERFORMANCE METRICS:")
print(f" Model Accuracy: {agentic_metrics['accuracy']:.4f}")
print(f" Model AUC: {agentic_metrics['auc']:.4f}")
print(f" Average Precision: {agentic_metrics['avg_precision']:.4f}")
print(f" Brier Score: {agentic_metrics['brier_score']:.4f}")
print(f" Average Uncertainty: {agentic_metrics['avg_uncertainty']:.4f}")

# Advanced capabilities
print("\nADVANCED CAPABILITIES IMPLEMENTED:")
print(" Multi-head Attention for Feature Interactions")
print(" Economic Cycle Integration")
print(" Uncertainty Quantification (Epistemic + Aleatoric)")
print(" Counterfactual Explanations")
print(" Feature Importance Analysis")
print(" Economic Sensitivity Analysis")
print(" Adversarial Training for Robustness")
print(" Comprehensive Risk Reporting")
print(" Advanced Data Generation with Economic Cycles")

print("\n" + "="*80)
print("AGENTIC AI CREDIT RISK ASSESSMENT SUCCESSFULLY COMPLETED")
print("="*80)
```
